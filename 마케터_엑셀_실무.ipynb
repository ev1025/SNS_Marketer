**엑셀 기본옵션 설정하기**
1. 엑셀 켰을 때 기본 설정
    - 옵션 - 일반 - 새통합 문서 만들기
    - 내가 저장해둔 글꼴을 불러올 수도있고 시트 수, 크기도 설정가능(크기는 11이 적당)
2. 엑셀 켰을 때 바로 시트화면
    - 옵션 - 일반 - 시작옵션(맨아래)
    - '이 응용프로그램 시작할때 ~' 체크해제 하면 엑셀 켤 때 바로 시트 화면으로 나옴
3. 한/영 자동으로 바뀌지 않게 하기
    - 옵션 - 언어 교정 - 자동고침 옵션 -자동고침 - 한영자동고침 해제
4. 자동저장 옵션
    - 옵션 - 저장 - 자동 복구 정보 저장 간격 - 짧게 설정
    - 옵션 - 저장 - 기본적으로 내 컴퓨터에 저장
5. 자동채움 목록만들기(월,화,수,목,금,토,일)
    - 옵션 - 고급 - 일반 - 사용자지정목록편집버튼
6. 개체 선택을 쉽게 할 수 있도록 하는 명령어
    - 옵션 - 빠른실행도구모음 - 명령선택(모든 명령) - 개체선택 추가
7. 빠른실행도구모음에 추가하는 방법
    - 옵션 - 빠른실행도구모음 - 사용자지정(오른쪽아래) - 가져오기/보내기(도구모음 창)
    - 리본메뉴에서 우클릭하고 추가
8. 옵션 - 리본설정 - 개발자도구

**명령어 단축키**
- 수식확인 : 컨트롤  + ~
- 오늘 날짜 : 컨트롤 + ;
- 현재 시간 : 컨+쉬+;
- 셀 서식 : 컨트롤 + 1
- 사용자형식 숫자 한글로 읽기 : [DBNum4]

- 분수 입력 : 0띄고 1/4 = $\frac{1}{4}$

**자동채우기**
- 문자 : 같은 문자 채움
- 숫자
    - 같은 것 복사 : 하나 땡기기
    - 연속형 복사 : 컨트롤 땡기기
    - 패턴 복사 : 두 셀 이상 선택해서 땡기기
- 날짜 / 시간(시간단위)
    - 같은것 복사 : 컨트롤 땡기기
    - 연속형 복사 : 그냥 땡기기
    - 패턴복사 : 두 셀이상
- 문자 + 숫자
    - 숫자만 변경됨
    - 숫자가 여러개면 마지막 숫자만 변경

- 자동 채우기 : 컨트롤 + e (셀병합이 있을때는 안됨)
    - 합정동 셀 누르고 컨트롤 + E 하면 밑으로 데이터 채워줌
    - 송산동, 평리동은 채워지고 무악동은 셀병합되어 있어서 안채워짐
