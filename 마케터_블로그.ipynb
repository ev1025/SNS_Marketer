### 네이버 스마트플레이스

- 아이디는 회사나 대표자 아이디로
- 셋팅은 PC로 관리는 모바일로
- 권한 가져오려면
    - 메인 가운데 '업체찾기' - '주인변경'
    
**기본정보탭**
- 신규등록(신규업체 버튼)
    - 업체명(사업자등록증 이름대로) : 이름 다르게 하면 지도에서 사라짐
    - 업체사진
        - 회사 로고 넣지 말기 (네이버는 상품이나 풍경사진이 올라온걸 좋아함)
        - 이미지 파일명을 내가 원하는 검색어 키워드로 바꾸고 띄어쓰기 하지말기
        - 최근 사진으로 올리는게 좋다.(다른 이름으로 저장해서 날짜 변경되게)
        - 고화질 사진을 올리는게 좋다.
    - 상세설명 : 네이버는 키워드 기반으로 움직이기 때문에 키워드를 잘 입력하는게 좋음
    - 대표키워드
        - 지역명 삽입(그래야 경쟁이 덜 됨)
        - 노출되지 않는 단어들은 과감히 삭제하기
    - 전화번호
        - 네이버는 지역전화번호를 입력하면 더 부가점을 준다.
        - 스마트콜 사용하기(네이버에서 신경을 많이 쓰고 있다. -> 성과측정가능)
        - 스마트콜을 채널별로 번호를 다르게 설정해놓고 확인해보면 광고 효과를 분석할 수 있다.(인스타, 현수막, 네이버 등)
    - 찾아오시는 길 설명
        - 도보거리, 차량거리,주차




**부가정보, 가격정보**
- 채널 모두 기재
- 가격표 사진
    - 파일명 : 업체명가격표
    - 가격정보 추가

**휴무일,영업시간**

- 같은 폰으로 사진 여러장 찍어서 나누면 안좋음

**네이버 블로그리뷰**

- 우리 회사 블로그 쓸 때 : 블로그 내용에 지도로 위치태그 X, 네이버지도 공유버튼으로 복사한 URL을 내용에 넣기

**상위노출(네이버에서 발표한 상위노출방법)**
- 네이버톡톡
    - 마케팅 메세지 : 무료로 광고 보낼 수 있다.
    - 네이버 쿠폰과 연계할 것
    - 네이버 지도 저장하면 쿠폰 주기?

- 네이버주문
- 네이버예약

    - 첫번째 노출될 경우만 0.5원
    - 무조건하자

- 네이버스마트플레이스

### 네이버 블로그

- 절대 외부에서 글쓰지말것(같은 ip로 많은 글이 올라올 경우 좋지않음)

- 1,000자 이상 텍스트(상위노출 1,500자 이상)
- 법인 회사는 공식블로그 인증 신청할 수 있다.
- 정보성 글이 노출 올리기 좋다.

    - 포스트 수
    - 쓰기 빈도(월, 목, 월, 목 이런식으로 쓰는게 좋음)

    - 방문자수(고유), 방문수(몇 번 왔는지), 페이지뷰(게시글 몇개 봤는지, 시리즈물 올리기(밑에 링크 올리기)), **이웃 수**, **스크랩 수**
