- 케러셀 형식 : 사진을 여러장 올리는 것
- 인스타그램의 성과는 **도달**
- 단컷은 한 번 보고 넘어가면 다시 나오지 않는데, **케러셀(Carousel)** 형식은 한 장만 보고 넘어갈 경우 다음 장으로 다시 노출됨
- 프로페셔널 비즈니스 계정(기업)
- 프로페셔널 크리에이터 계정(내가 하고싶으면)
- 인사이트 - 팔로워 - 내 팔로워가 가장 많이 이용하는 시간에 컨텐츠 올리기
- 인스타 추천(벤치마킹)
    - 빙그레, 키크니
    - 갤럭시, 팬톤
    - 오늘의 집, 이니스프리
- 프로필
    - 320x 320
    - 링크 5개 넣을 수 있음(오시는길도 넣기)
    - 이름에 핵심키워드 넣기
    - 소개 태그에 이모지 넣기
    - 인스타는 프로필에 줄바꿈이 안되서 **인스타공백닷컴**에서 변경해서 사용할것
    - 글 쓸 대도 인스타공백닷컴에서 첫줄 공백만들고 쓰면 아이디 밑줄부터 글써짐
- 스토리 + 하이라이트
    - 세일내역, 찾아오는길, 후기
    - 최신 스토리가 올라오면 하이라이트가 왼쪽으로 넘어감
    - 스토리 - 위에 사람모양 - 링크 - URL, 밑에 스티커택스트 변경하면 링크이름 설정가능

- 릴스
    - 오디오 쓰고, 주제추가 꼭 넣어야 도달률이 높아짐
    - 비즈니스계정은 오디오 선정이 한정(저작권)되어 있어서 탬플릿이나, 올라온 릴스 중에 확인해서 오디오 사용 누르고 이용하기
    - 어플사용 : 릴스앱(아이폰), 캡컷
    - 인스타그램 콜랍스 : 공동 작업자 1인을 초대하여 두 인스타계정에 동시에 올라가고 좋아요 합산됨
        - 게시글 쓸 때 **사람 태그하기**
        - 인플루언서가 태그해주면 나한테 이득
- 소셜 큐레이션(=알고리즘)
- 피드
    - 필터기능을 사용해서 톤의 통일감을 맞추기
    - 3의 법칙
    - 바둑판 레이아웃
    - 손으로 물건 들고 찍기같은 컨셉을 잡아도 좋음
    - 사진 위아래에 흰색 여백을 둬서 피드 사이의 간격을 줄 수 있음
- 캡션(글씨 쓰는 영역)
    - 인스타공백닷컴
    - 해시태그 10~15개가 가장 효율이 좋다.
        - 대댓글에 넣기
        - 캡션에 3\~5개, 댓글에 5~7개가 가장 좋다(브랜드 해시태그2개정도 나머진 글에 맞는 해시태그)
        - 미디언스(해시태그 연관분석) - 인스타 사용시간도 조회가능
        - 태그야놀자(안드로이드앱)
        - 컨셉해시태그 : 52소주, 린자g
        - 우리 고객을 위한 해시태그 만들기
    
    - 페이스북(미러링)으로 같은 글을 올릴 수 있는데, 대댓글에 해시태그를 넣으면 페이스북에 안따라가므로, 캡션중간중간에 해시태그를 넣어야한다.

- 인스타그램 인게이지먼트(블로그지수같은거, 네이버검색)
    - 팔로우 대비 좋아요 5~10%가 안되면 돈주고 만든 좋아요
- 인스타그램 위치추가
    - 페이스북에 체크인에 위치추가 하고 1주일정도 지나야 인스타그램에 생김
    - 위치사용 자주 안하면 위치 삭제
