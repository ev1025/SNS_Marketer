**유튜브 활동**
- 콘텐츠 제공자 : 개인, 회사, 기관 등 다양한 기관이 영상 콘텐츠를 제작하고 업로드합니다.
- 사용자(시청자) : 전 세계의 인터넷 사용자는 유튜브에서 업로드 된 많은 영상을 상호작용합니다.
- 광고주 : 기업 및 개인 광고주는 유튜브에 광고를 게재하여 상품이나 서비스를 홍보하고 판매를 늘립니다.광고주는 광고게재를 통해 수익을 창출하고 유튜브는 광고 게재 수수료를 얻습니다.


- UGC(User, Generated Content) : 일반인들이 자발적으로 정보전달성으로 상품리뷰를 하는것

**광고 종류**

- 직접광고 : 광고주를 위해 제작된 콘텐츠
- 스폰서십 : 콘텐츠 제작에 일부 비용을 받음

**페이스북 머신러닝에 최적화된 캠페인**
- 신규고객 전환 캠페인
- 기존구매 고객전환률 높이는 캠페인

**페이스북 DCO**
- 광고영상 10개 X 텍스트5개 X 제목 5개, 설명5개, Call To Action(더 알아보기) 5개를 만들어서 광고 송출
- 일주일에 하위 50%씩 하위 광고 제거하면 광고성과가 올라감

**인플루언서 마케팅**
- 로고스(10%논리) / 파토스(30% 감성) / 에토스(60% 관계/신뢰)
- 장점
    - 특정 매니아 고객 타게팅, 외주화
    - 매력산업 : 높은 구매전환율
    - Paid + 브랜디드 마케팅을 동시에 하게됨
- 예산의 2~40%정도를 사용하는게 좋음
- 짧게 걸릴 수록 좋음(최대한 6주를 넘기지 않는게 좋음)
- 인플루언서 리스트업
    - 콘텐츠 퀄리티
    - 콘텐츠 카테고리
    - 스토리텔링
    - 팔로워 퀄리티
    - 정치색
    - 논란 유무 체크 (oo 논란)
    - 영상 5개 이상 꼭 보기(구독자를 부르는 호칭을 알아내서 대화할 때 얘기함)
        - 해당 인플루언서 영상에 좋아요를 다 눌러놓으면 관련 영상이 나옴
    - 유튜브 정보에 e메일로 보내면 됨
- 릴리즈(광고시행)
    - 초반의 여론이 컨텐츠의 이미지를 결정하기 때문에 초반 댓글작업 필요
    - 우리 홈페이지등으로 랜딩시킬 추적가능한 링크(UTM)
    - 유입을 도울 썸네일
    - 제목키워드 딱 1개만
    - 루즈한 부분은 브랜드 어필포인트여도 다 삭제
- 인플루언서 추천 규모
    - 마이크로인플루언서 : 소수 소속사, 마이너& 특정 주제 전문가 가능성 높음, 성장성 높음
    - 나노 인플루언서 : 조회수기대보단 여론 조성용, 뜰유투버찾기, 3B 추천
    - 팔로워 대비 좋아요 댓글수, 구매력 있는 팔로워인지 확인

- 2차 라이선스(광고영상을 타 매체에 이용할 권리)
    - 링크활용
    - 편집 및 단순 업로드
    - 편집 및 광고매체 집행
    - 초상권 및 성명권
    - 처음에 계약할 때 2차라이선스 비용까지 포함해서 계약하면 돈을 아낄 수 있다.
    - 1~3개월이 표준이고, 잘하면 무기한으로 이용할 수 있다.(초상권 등의 문제)
