**유튜브 활동**
- 콘텐츠 제공자 : 개인, 회사, 기관 등 다양한 기관이 영상 콘텐츠를 제작하고 업로드합니다.
- 사용자(시청자) : 전 세계의 인터넷 사용자는 유튜브에서 업로드 된 많은 영상을 상호작용합니다.
- 광고주 : 기업 및 개인 광고주는 유튜브에 광고를 게재하여 상품이나 서비스를 홍보하고 판매를 늘립니다.광고주는 광고게재를 통해 수익을 창출하고 유튜브는 광고 게재 수수료를 얻습니다.


- UGC(User, Generated Content) : 일반인들이 자발적으로 정보전달성으로 상품리뷰를 하는것

**광고에서 머신러닝이 돌아가는 건 수**
- 구글AD = 300건(한달)
- 페이스북 50건(일주일)

**페이스북 머신러닝에 최적화된 캠페인**
- 신규고객 전환 캠페인
- 기존구매 고객전환률 높이는 캠페인

**페이스북 DCO**
- 광고영상 10개 X 텍스트5개 X 제목 5개, 설명5개, Call To Action(더 알아보기) 5개를 만들어서 광고 송출
- 일주일에 하위 50%씩 하위 광고 제거하면 광고성과가 올라감

**인플루언서 마케팅**
- 로고스(10%논리) / 파토스(30% 감성) / 에토스(60% 관계/신뢰)
- 장점
    - 특정 매니아 고객 타게팅, 외주화
    - 매력산업 : 높은 구매전환율
    - Paid + 브랜디드 마케팅을 동시에 하게됨
- 예산의 2~40%정도를 사용하는게 좋음
- 짧게 걸릴 수록 좋음(최대한 6주를 넘기지 않는게 좋음)

**북국성지표(NSM)**
- 고객이 제품/서비스에서 얻는 가치
- 지표가 제품의 성과를 보여야한다.
- 대부분의 부서를 커버해야 한다.
- NSM이 올랐는데 매출이 그대로다? -> 제품을 변경해야한다.

**가이드**
- 소구점
    - 고객의 행동과 동기 관점
    - 오랜 관찰, 인사이트뽑기
